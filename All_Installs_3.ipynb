```python
def printTime(strt, fin):
    '''print execution times '''
    totalTime = fin - strt
    totalTime= round(totalTime, 0)
    if totalTime < 60.0:
        print(f'completed: {totalTime} second(s)')
    elif totalTime >= 60.0:
        minutes = totalTime//60
        seconds = totalTime - (minutes * 60)
        seconds = round(seconds, 2)
        print(f'completed: {minutes} minutes {seconds} second(s)')

def getTensorflow():
    %pip install -q tensorflow
    print(f'installed: {C.BIBlue}tensorflow{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}tensorflow{C.ColorOff}'

def getTensorflowDatasets():
    %pip install -q tensorflow-datasets
    print(f'installed: {C.BIBlue}tensorflow-datasets{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}tensorflow-datasets{C.ColorOff}'

def getTFLiteModMaker():
    %pip install -q tflite-model-maker
    print(f'installed: {C.BIBlue}tflite-model-maker{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}tflite-model-maker{C.ColorOff}'

def getTFHub():
    %pip install -q --upgrade tensorflow-hub
    print(f'installed: {C.BIBlue}tensorflow-hub{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}tensorflow-hub{C.ColorOff}'
    
# def getTFLite():
    # %pip install -q tflite-model-maker
    # print(f'installed: {C.BIBlue}tflite-model-maker{C.ColorOff}')
    # sleep(1)
    # return f'installed: {C.BIBlue}tflite-model-maker{C.ColorOff}'

def getNumpy():
    %pip install -q numpy
    print(f'installed: {C.BIBlue}numpy{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}numpy{C.ColorOff}'
    
def getCV2():
    %pip install -q opencv-python-headless # works.
    print(f'installed: {C.BIBlue}opencv-headless{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}opencv-headless{C.ColorOff}'
    # %pip install -q opencv-contrib-python-headless

def getMatplotlib():
    %pip install -q matplotlib
    print(f'installed: {C.BIBlue}matplotlib{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}matplotlib{C.ColorOff}'
    
def getSciPi():
    %pip install -q SciPy
    print(f'installed: {C.BIBlue}SciPy{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}SciPy{C.ColorOff}'

def getSeaborn():
    %pip install -q seaborn
    print(f'installed: {C.BIBlue}seaborn{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}seaborn{C.ColorOff}'

def getResults():
    %pip install -q --upgrade results
    print(f'installed: {C.BIBlue}results{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}results{C.ColorOff}'

def getBeautifulSoup():
    ''' '''
    %pip install -q beautifulsoup4
    print(f'installed: {C.BIBlue}beautifulsoup4{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}beautifulsoup4{C.ColorOff}'

def get_htmlib():
    '''USED WITH BeautifulSoup'''
    %pip install -q html5lib
    print(f'installed: {C.BIBlue}html5lib{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}html5lib{C.ColorOff}'

def get_lxml():
    '''USED WITH BeautifulSoup'''
    %pip install -q lxml
    print(f'installed: {C.BIBlue}lxml{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}lxml{C.ColorOff}'

def getImutils():
    '''Image processing utility '''
    %pip install -q imutils
    print(f'installed: {C.BIBlue}imutils{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}imutils{C.ColorOff}'
    
def getPillow():
    '''Image processing utility '''
    %pip install -q pillow
    print(f'installed: {C.BIBlue}pillow{C.ColorOff}')
    # sleep(1)
    return f'installed: {C.BIBlue}pillow{C.ColorOff}'
    
print('ok')
```
