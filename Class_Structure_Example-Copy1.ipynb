```python
# https://stackoverflow.com/questions/222877/what-does-super-do-ion-python-difference-between-super-init-and-expl

from __future__ import absolute_import, unicode_literals
from BashColors import C
import json, os, pprint, sys

class A(object):  # or class A:
    def __init__(self):
        self.contentPath = None
        print("Class: A")
        
    
class B(A):
    def __init__(self):
        super().__init__()  # or super(B, self).__init_()
        print("Class: B")
   
class C(A):
    def __init__(self):
        super().__init__()
        print("Class: C")
  
class D(B, C):
    def __init__(self):
        super().__init__()
        print("Class: D")
    
d = D()
```
